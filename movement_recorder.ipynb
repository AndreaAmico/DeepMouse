```python
import win32gui
import nptable
import time
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

```python
def progress_bar(current_value, max_value):
    progress = ((current_value+1)/max_value)*100
    if progress>98: progress=100
    print('\r[{0}{1}] {2:.1f}%'.format('#'*int(progress/2), ' '*(50-int(progress/2)), progress), end='')
```

```python
N = 500
t = []
with open('./data/test.txt', 'a') as f:
    for index in range(N):
        x, y = win32gui.GetCursorPos()
        f.write('{},{}\n'.format(x, y))
        t.append(time.time())
        time.sleep(0.01)
        progress_bar(index, N)
```

```python
diff_t = np.diff(t)

fig, ax = plt.subplots(1, figsize=(6,4))

ax.plot(diff_t*1000)
ax.set_xlabel('Acquisition-index')
ax.set_ylabel('Time difference (ms)')
ax.set_title('Data acquisition stability')
ax.set_xlim(0, 500)

fig.savefig('./plots/data_acquisition_stability2.png', bbox_inches='tight')
```
